### Exercici 2

**Puntuació i validació**: Aquest exercici val 2 punts només si passa tots els tests, si només passa algun dels tests la nota serà 0.

Crea un sistema bàsic de gestió de vehicles que utilitza herència simple per organitzar diferents tipus de vehicles.

Classe Base **Vehicle**:

&nbsp;&nbsp; Atributs: marca (String), model (String), i any (int).

&nbsp;&nbsp; Constructor que inicialitza tots els atributs.

&nbsp;&nbsp; Getters i Setters necessaris

&nbsp;&nbsp; Override del mètode d'imprimir cadena que imprimeix la informació bàsica del vehicle. Segons aquest format:

&nbsp;&nbsp;&nbsp;&nbsp; *"Marca: " + marca + ", Model: " + model + ", Any: " + any + "\n"*


Classe **Cotxe**:

&nbsp;&nbsp; Atributs addicionals: numPortes (int).

&nbsp;&nbsp; Constructor que inclou els atributs de la classe base i el nou atribut.

&nbsp;&nbsp; Getters i Setters necessaris

&nbsp;&nbsp; Override del mètode d'imprimir cadena per incloure numPortes a la informació mostrada. Segons aquest format:&ens

&nbsp;&nbsp;&nbsp;&nbsp; *"Nombre de portes: " + numPortes*


Classe **Moto**:

&nbsp;&nbsp; Atributs addicionals: cilindrada (int).

&nbsp;&nbsp; Constructor que inclou els atributs de la classe base i el nou atribut.

&nbsp;&nbsp; Getters i Setters necessaris

&nbsp;&nbsp; Override del mètode d'imprimir cadena per incloure cilindrada a la informació mostrada. Segons aquest format:

&nbsp;&nbsp;&nbsp;&nbsp; *"Cilindrada: " + cilindrada + "cc"*

Prova-ho amb aquest main:

```java
public class Main {
    public static void main(String[] args) {
       
        Cotxe cotxe = new Cotxe("Toyota", "Corolla", 2020, 4);
        Moto moto = new Moto("Honda", "CBR", 2021, 650);

        System.out.println("Informació del Cotxe:");
        System.out.println(cotxe);

        System.out.println("\nInformació de la Moto:");
        System.out.println(moto);
    }
}
```

Exemple de sortida:

```text
Informació del Cotxe:
Marca: Toyota, Model: Corolla, Any: 2020
Nombre de portes: 4

Informació de la Moto:
Marca: Honda, Model: CBR, Any: 2021
Cilindrada: 650cc
```
