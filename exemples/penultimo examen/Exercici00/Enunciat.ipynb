### Exercici 0

**Puntuació i validació**: Aquest exercici val 2 punts només si passa tots els tests, si només passa algun dels tests la nota serà 0.

Crea un programa en Java que realitzi les següents tasques:

* La funció 'ésPrimer' que rep un enter com a paràmetre i retorna cert si és un número primer.

* La funció 'inicialitzarLlista' que rep dos paràmetres enters 'total' i 'numPrimers'. Aleshors crea un ArrayList d'enters menors de 100 amb un total de 'total' números enters, d'aquests hi ha almenys, la quantitat de números primers definia a 'numPrimers'.

* La funció 'cercaPrimers' que rep com a paràmetre un ArrayList d'enters i retorna l'ArrayList amb els números primers de la llista rebuda.

**Nota:** Decideix la visibilitat de les funcions, per tal de passar els testos.

Pistes:

* Per determinar si un número és primer, pots crear un mètode que iteri des de 2 fins a l'arrel quadrada del número en qüestió. Si el número és divisible per qualsevol número en aquest rang, llavors no és primer. És a dir:

```text
- si el número és menor o igual a 1, retorna false
- si el número és més gran que 1:
    * calcular l'arrel quadrada del número
    * fer un bucle des de 2 fins a l'arrel quadrada del número (inclòs), i dins del bucle:
      si el mòdul del número amb el comptador del bucle dóna 0, retornar false
- si el bucle acaba, retornar true
```

* Recorda utilitzar bucles, com for o while, per iterar sobre els elements de la llista i també per verificar si un número és primer.
Usa condicions, com if o if-else, per implementar la lògica de decisió.

Prova-ho amb aquest main:

```java
public class Main {
    public static void main(String[] args) {
        ArrayList<Integer> numeros = inicialitzarLlista(20, 10);
        ArrayList<Integer> primers = cercaPrimers(numeros);
        
        System.out.println("Llista de números: " + numeros);
        System.out.println("Números primers trobats: " + primers);
    }
}
```

Exemple de sortida:

```text
Llista de números: [73, 23, 71, 2, 19, 83, 11, 67, 41, 5, 28, 24, 18, 48, 29, 45, 46, 33, 1, 3]
Números primers trobats: [73, 23, 71, 2, 19, 83, 11, 67, 41, 5, 29, 3]
```
