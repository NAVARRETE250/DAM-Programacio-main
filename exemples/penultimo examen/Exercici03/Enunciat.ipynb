### Exercici 3

**Puntuació i validació**: Aquest exercici val 2 punts només si passa tots els tests, si només passa algun dels tests la nota serà 0.

Crea un sistema de notificacions on diferents tipus de notificacions.

Classe **Notificacio**:

&nbsp;&nbsp; Defineix una propietat destinatari de tipus String.

&nbsp;&nbsp; Defineix un atribut prioritat (tipus int), que determina l'ordre en què les notificacions haurien de ser processades.

&nbsp;&nbsp; Getters i Setters

&nbsp;&nbsp; Constructor que inicialitza el destinatari.

&nbsp;&nbsp; Mètode enviaNotificacio(): no tindrà implementació en aquesta classe.

&nbsp;&nbsp; Mètode ordenaPrioritat: que rep com a paràmetre una llista de notificacions i ordena la llista segons prioritat. S'ha de poder cridar sense instancia.


Classe **NotificacioEmail**:

&nbsp;&nbsp; Implementa el mètode enviaNotificacio() per mostrar un missatge que digui "Enviant Email a [destinatari]".


Classe **NotificacioPush**:

&nbsp;&nbsp; Implementa el mètode enviaNotificacio() per mostrar un missatge que digui "Enviant Notificació Push a [destinatari]".


Classe **NotificacioSMS**:

&nbsp;&nbsp;  Implementa el mètode enviaNotificacio() per mostrar un missatge que digui "Enviant SMS a [destinatari]".


**Nota:** Decideix les propietats i herència de cada classe, per tal de passar els testos.

Prova-ho amb aquest main:

```java
public class Main {
    public static void main(String[] args) {
        List<Notificacio> notificacions = new ArrayList<>();
        notificacions.add(new NotificacioEmail("usuari@example.com", 2));
        notificacions.add(new NotificacioSMS("123456789", 1));
        notificacions.add(new NotificacioPush("usuariPush", 3));
        notificacions.add(new NotificacioSMS("987654321", 0));
        notificacions.add(new NotificacioEmail("usuari@ieti.cat", 4));

        Notificacio.ordenaPerPrioritat(notificacions);

        for (Notificacio notificacio : notificacions) {
            notificacio.enviaNotificacio();
        }
    }
}
```

Exemple de sortida:

```text
Enviant SMS a 987654321
Enviant SMS a 123456789
Enviant Email a usuari@example.com
Enviant Notificació Push a usuariPush
Enviant Email a usuari@ieti.cat
```
