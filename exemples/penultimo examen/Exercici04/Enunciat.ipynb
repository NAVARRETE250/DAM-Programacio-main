### Exercici 4

**Puntuació i validació**: Aquest exercici val 2 punts només si passa tots els tests, si només passa algun dels tests la nota serà 0.

Crea un sistema de gestió d'inventari que permeti afegir, gestionar i mostrar informació sobre diferents tipus de productes en una botiga. 

Classe **Producte**:

&nbsp;&nbsp; Atributs: nom (String), preu (double).

&nbsp;&nbsp; Constructor que inicialitza els atributs.

&nbsp;&nbsp; Getters i Setters per a cada atribut

&nbsp;&nbsp; Sobreescriptura del mètode per la informació bàsica del producte, segons aquest estil:

&nbsp;&nbsp;&nbsp;&nbsp; "Nom: " + nom + ", Preu: " + preu + "€"


Classe **Electronica**:

&nbsp;&nbsp; Atributs addicionals: garantia (int, nombre d'anys de garantia).

&nbsp;&nbsp; Constructor que inclou tots els atributs.

&nbsp;&nbsp; Getters i Setters per garantia

&nbsp;&nbsp; Sobreescriptura del mètode per la informació bàsica del producte, segons aquest estil:

&nbsp;&nbsp;&nbsp;&nbsp; ", Garantia: " + garantia + " anys"


Classe **Alimentacio**:

&nbsp;&nbsp; Atributs addicionals: dataCaducitat (String).

&nbsp;&nbsp; Constructor que inclou els atributs de la classe base i el nou atribut.

&nbsp;&nbsp; Getters i Setters per dataCaducitat

&nbsp;&nbsp; Sobreescriptura del mètode per la informació bàsica del producte, segons aquest estil:

&nbsp;&nbsp;&nbsp;&nbsp; ", Data de caducitat: " + dataCaducitat


Classe **Inventari**:

Atributs: productes (Llista de Producte).

&nbsp;&nbsp; Funció afegeixProducte(Producte producte)

&nbsp;&nbsp; mostraInventari() que itera sobre la llista de productes i imprimeix la informació de cada producte.

&nbsp;&nbsp;&nbsp;&nbsp; 


Prova-ho amb aquest main:

```java
public class Main {
    public static void main(String[] args) {
       
        Inventari inventari = new Inventari();

        inventari.afegeixProducte(new Electronica("Televisor", 599.99, 2));
        inventari.afegeixProducte(new Alimentacio("Pa", 1.50, "2024-05-01"));
        inventari.afegeixProducte(new Electronica("Smartphone", 999.99, 1));
        inventari.afegeixProducte(new Alimentacio("Llet", 0.99, "2023-12-31"));

        System.out.println(inventari);
    }
}
```

Exemple de sortida:

```text
Nom: Televisor, Preu: 599.99€, Garantia: 2 anys
Nom: Pa, Preu: 1.5€, Data de caducitat: 2024-05-01
Nom: Smartphone, Preu: 999.99€, Garantia: 1 anys
Nom: Llet, Preu: 0.99€, Data de caducitat: 2023-12-31
```
